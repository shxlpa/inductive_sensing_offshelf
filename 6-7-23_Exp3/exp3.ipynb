In this experiment, we used the LCR meter to see what order of magnitude it placed the inductance in, and also to see if it empirically verified our trends.

We ran 31 trials at {100 Hz, 1 kHz, 10 kHz, 100 kHz}, delivering 1 Vrms voltage.

We found that there was a frequency dependence in the inductance, possibly due to Eddy Currents in the magnet.

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
gap_distance = []
gap = 2.5
for i in range(31):
    gap_distance += [gap]
    gap+=1
print(len(gap_distance))
```
