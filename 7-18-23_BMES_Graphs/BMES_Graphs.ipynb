# Putting all the figures in one
- Normalize Axes
- Start them after 9.5
- Error bars (do after formatting)

```python
def meanstd(trial0, trial1, trial2, trial3, trial4):
    mean = np.mean( np.array([ trial0, trial1, trial2, trial3, trial4 ]), axis=0 )
    # print(mean)
    
    stdev = np.std( np.array([ trial0, trial1, trial2, trial3, trial4 ]), axis=0 )
    # print(stdev)
    
    #2x st dev
    error = []
    i = 0
    for x in range(12):
        error += [stdev[i]*2]
        i += 1
    
    return mean, error
```
